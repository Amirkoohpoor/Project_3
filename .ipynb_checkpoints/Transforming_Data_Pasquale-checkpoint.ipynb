```python
# parse the address string on the left of "|" which gives us the Street Number and Name
realtor['Street']=realtor['Address'].str.split('|').str[0]
#parse the address string on the right of "|" which gives us the City and Province and Postal code
realtor['City - Province']=realtor['Address'].str.split('|').str[1]
#drop the postal code from teh above string
realtor['City Province']=realtor['City - Province'].str[:-6]
# parse the City Province sttring on the left of ", which gives us the City Name
realtor['City']=realtor['City Province'].str.split(',').str[0]
# parse the City Province sttring on the right  of ", which gives us the Province 
realtor['Province']=realtor['City Province'].str.split(',').str[1]
# parse the last 6 digits of the afull address to capture the Postal code
realtor['Postal_Code']=realtor['Address'].str[-6:]
#print the DataFrame
realtor.head()
```

```python
#drop the Column 0 -"Full Address", Column 10 = City-Province and Column 11 - City Province
df = realtor.drop(realtor.columns[[0,5,6]], axis=1)  
df=df.dropna()
# rint the new DataFRame
df
```

```python
#Import the Zolo csv file
zolo = pd.read_csv('C:\\Users\\pasqu\\OneDrive\\Documents\\Pasquale Github\\Project_3\\output_1.csv')
zolo
```

```python
# new data frame with split value columns 
new = zolo["Days on Site"].str.split("(", n = 1, expand = True) 
# making separate Days column from new data frame 
zolo["Days"]= new[0] 
# making separate Date posted on MLScolumn from new data frame 
zolo["Date posted on MLS"]= new[1].str[:-1] 
#house_df["Date posted on MLS"].str[:-1]
# Dropping old Name columns 
zolo.drop(columns =["Days on Site","Days","address","city"], inplace = True) 
#rename the MLS number column
zolo.rename(columns={'MLS®#':'Mls Number'}, inplace=True)
# drop columns
zolo_clean_df = zolo.drop(zolo.columns[[0]], axis=1)   
# df display 
zolo_clean_df
```

```python
merge_table = pd.merge(df,zolo_clean_df, on="Mls Number")
merge_table.dropna()  
merge_table.drop_duplicates
merge_table
```
